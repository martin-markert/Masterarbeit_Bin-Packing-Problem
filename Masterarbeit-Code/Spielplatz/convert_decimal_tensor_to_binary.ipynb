```python
def binary_encoder(x,binary_dim):
    # x: size x 1
    if binary_dim == 1:
        x_encoding = x
    else:
        binary_list = []
        divide = x
        for i in range(binary_dim):
            binary = divide % 2
            binary_list.insert(0,binary)
            divide = torch.div(divide,2,rounding_mode='trunc')

        x_encoding = torch.stack(binary_list,-1).flatten(-2,-1)
    return x_encoding


def binary_encoder_2(x, binary_dim):
    # x: Tensor der Form (size, 1)
    if binary_dim == 1:
        return x
    x = x.to(torch.int64)
    # Erzeuge Exponenten für die Bits: 2^(binary_dim-1), ..., 2^0
    exponents = 2 ** torch.arange(binary_dim - 1, -1, -1, device=x.device)
    # Division und Modulo für jedes Bit
    x_encoding = (x // exponents) % 2
    return x_encoding


import torch
import warnings

def dec2bin(x, bits):
    max_val = 2**bits - 1
    if (x > max_val).any():
        warnings.warn(
            f"Bei dec2bin() werden Werte in x werden auf {bits} Bits abgeschnitten. Maximal darstellbarer dezimaler Wert: {max_val}", 
            UserWarning
        )
    
    mask = 2 ** torch.arange(bits - 1, -1, -1, device=x.device, dtype=x.dtype)
    return (x & mask).ne(0).int()


# def dec2bin(x, bits):
#     mask = 2 ** torch.arange(bits - 1, -1, -1, device=x.device, dtype=x.dtype)
#     return (x & mask).ne(0).int()





x = torch.tensor([[0], [5], [7], [90], [-12]], dtype=torch.int32)
binary_dim = 20

print(binary_encoder(x,binary_dim))
print("\n")
# print(binary_encoder_2(x,binary_dim))
print(dec2bin(x, binary_dim))
```
